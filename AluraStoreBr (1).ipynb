### Importação dos dados



```python
import pandas as pd
import matplotlib.pyplot as plt


url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento


```python
faturamento = loja['Preço'].sum()
faturamento2 = loja2['Preço'].sum()
faturamento3 = loja3['Preço'].sum()
faturamento4 = loja4['Preço'].sum()

print(f'O faturamento total da Loja 1 é {faturamento}')
print(f'O faturamento total da Loja 2 é {faturamento2}')
print(f'O faturamento total da Loja 3 é {faturamento3}')
print(f'O faturamento total da Loja 4 é {faturamento4}')
```

```python
lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']
valores = [faturamento, faturamento2, faturamento3, faturamento4]

plt.figure(figsize=(7, 7))
plt.pie(valores, labels=lojas, autopct='%1.1f%%', startangle=140)
plt.title('Faturamento por Loja (%)')
plt.axis('equal')
plt.show()
```

# 2. Vendas por Categoria


```python
produtos = loja['Categoria do Produto'].value_counts()
produtos2 = loja2['Categoria do Produto'].value_counts()
produtos3 = loja3['Categoria do Produto'].value_counts()
produtos4 = loja4['Categoria do Produto'].value_counts()

print("Loja 1 - Categorias mais populares:")
print(produtos.head(3))

print("\nLoja 2 - Categorias mais populares:")
print(produtos2.head(3))

print("\nLoja 3 - Categorias mais populares:")
print(produtos3.head(3))

print("\nLoja 4 - Categorias mais populares:")
print(produtos4.head(3))
```

```python
categorias_loja1 = loja['Categoria do Produto'].value_counts()

plt.figure(figsize=(10, 6))
categorias_loja1.plot(kind='bar', color='teal')
plt.title('Categorias Mais Vendidas da Loja 1')
plt.ylabel('Quantidade Vendida')
plt.xlabel('Categoria do Produto')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

# 3. Média de Avaliação das Lojas

```python
media = loja['Avaliação da compra'].mean()
media2 = loja2['Avaliação da compra'].mean()
media3 = loja3['Avaliação da compra'].mean()
media4 = loja4['Avaliação da compra'].mean()

print(f'A média de avaliação da Loja 1 é {media}')
print(f'A média de avaliação da Loja 2 é {media2}')
print(f'A média de avaliação da Loja 3 é {media3}')
print(f'A média de avaliação da Loja 4 é {media4}')
```

```python
todas_lojas = pd.concat([loja, loja2, loja3, loja4])
todas_lojas['Data da Compra'] = pd.to_datetime(todas_lojas['Data da Compra'])

# Agrupa por mês e calcula a média de avaliação
avaliacao_por_data = todas_lojas.groupby(todas_lojas['Data da Compra'].dt.to_period('M'))['Avaliação da compra'].mean()
avaliacao_por_data.index = avaliacao_por_data.index.to_timestamp()  # Converte índice para datetime

# Calcula as médias de cada loja individualmente
media_loja1 = loja['Avaliação da compra'].mean()
media_loja2 = loja2['Avaliação da compra'].mean()
media_loja3 = loja3['Avaliação da compra'].mean()
media_loja4 = loja4['Avaliação da compra'].mean()
media_geral = todas_lojas['Avaliação da compra'].mean()

# Cria o gráfico
plt.figure(figsize=(12, 6))
plt.plot(avaliacao_por_data.index, avaliacao_por_data.values, marker='o', linestyle='-', label='Avaliação da compra')

# Adiciona as linhas de média
plt.axhline(media_geral, color='red', linestyle='--', linewidth=2, label='Média Geral')
plt.axhline(media_loja1, color='blue', linestyle=':', label='Média Loja 1')
plt.axhline(media_loja2, color='green', linestyle=':', label='Média Loja 2')
plt.axhline(media_loja3, color='orange', linestyle=':', label='Média Loja 3')
plt.axhline(media_loja4, color='purple', linestyle=':', label='Média Loja 4')

# Ajustes finais
plt.title('Avaliação Média Mensal dos Clientes')
plt.ylabel('Avaliação Média')
plt.xlabel('Data')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

# Relatório

**Introdução:**

O presente relatório tem como objetivo analisar o desempenho de quatro lojas com base em métricas comerciais e operacionais, visando recomendar qual delas o Senhor João deve vender. Foram considerados os seguintes critérios: faturamento total, categorias e produtos mais e menos vendidos, avaliação média dos clientes e custo médio de frete. As análises foram realizadas por meio de exploração e visualização de dados, permitindo uma comparação objetiva entre as unidades.

**Análise dos Indicadores**

**Faturamento Total:**

A Loja 1 apresenta o maior faturamento, totalizando 1.534.509,12 reais, seguida pela Loja 2 (1.488.459,06 reais), Loja 3 (1.464.025,03 reais) e Loja 4 (1.384.497,58 reais). Esse indicador evidencia o melhor desempenho financeiro da Loja 1.

**Categorias de Produtos Mais Populares:**

As categorias mais vendidas foram similares entre as quatro lojas: móveis, eletrônicos e brinquedos. A Loja 3 se destaca com o maior volume de vendas nessas categorias, especialmente em móveis (499 unidades) e eletrônicos (451 unidades), demonstrando forte aderência a itens de maior demanda.

**Avaliação Média dos Clientes:**

A Loja 3 também possui a melhor média de avaliação dos clientes (4,05), seguida pela Loja 2 (4,04), Loja 4 (4,00) e Loja 1 (3,98). Esse critério indica maior satisfação dos consumidores em relação à experiência de compra na Loja 3.

**Produtos Mais e Menos Vendidos:**

Cada loja apresentou variações nos produtos mais e menos vendidos. A Loja 1, apesar do maior faturamento, teve menor variedade entre os itens mais vendidos, com forte presença de eletrodomésticos. Por outro lado, a Loja 3 mostrou equilíbrio entre móveis e utensílios domésticos, enquanto a Loja 4 se destacou por incluir produtos digitais entre os mais vendidos, como o curso de Power BI.

**Custo Médio de Frete:**

A Loja 4 apresentou o menor custo médio de frete (31,28 reais), o que pode representar maior eficiência logística. As demais lojas registraram custos crescentes, com a Loja 1 apresentando o maior custo (34,69 reais).

**Conclusão:**

Embora a Loja 1 apresente o maior faturamento absoluto, ela também possui a menor média de avaliação dos clientes e o maior custo de frete. Em contrapartida, a Loja 3 reúne um bom desempenho em múltiplos aspectos: elevado faturamento, melhor média de avaliações, categorias com alto volume de vendas e equilíbrio na variedade de produtos vendidos. Além disso, seu custo de frete está abaixo da média geral.

Com base nas análises realizadas, recomenda-se que o Senhor João venda a Loja 1. Apesar de seu faturamento elevado, a loja apresenta desvantagens operacionais e de experiência do cliente, o que pode comprometer sua competitividade no médio e longo prazo. As demais lojas, especialmente a Loja 3, mostram-se mais consolidadas em termos de satisfação do cliente, controle logístico e equilíbrio de vendas, sendo mais indicadas para continuidade e expansão do negócio.
